```python
# Importing

import tensorflow as tf

import numpy as np

import matplotlib.pyplot as plt

import pandas as pd
```

```python
celsius_train = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit_train = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
for c,f in enumerate(celsius_train):
  print("Celsius: {} = Fahrenheit: {}".format(f, fahrenheit_train[c]))
```

```python
# Model with single dense layer

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```python
# Compile model

model.compile(loss="mean_squared_error", optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
hist = model.fit(celsius_train, fahrenheit_train, epochs=500, verbose=True)
print("Finished Training the Model")
```

```python
plt.title("Loss Progression")
plt.plot(hist.history["loss"])
plt.xlabel("Epoch Number")
plt.ylabel("Loss")
```

```python
print(model.predict([100.00]))
```

```python
new_model = tf.keras.Sequential()
new_model.add(tf.keras.layers.Dense(units=4, input_shape=[1]))
new_model.add(tf.keras.layers.Dense(units=4))
new_model.add(tf.keras.layers.Dense(units=1))

new_model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss= "mean_squared_error"
)

new_hist = new_model.fit(celsius_train, fahrenheit_train, epochs=500, verbose=False)
```

```python
plt.title("Loss Progression")
plt.plot(new_hist.history["loss"])
plt.xlabel("Epoch Number")
plt.ylabel("Loss")
```

```python
print(new_model.predict([100]))
```
